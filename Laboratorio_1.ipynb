```python
##PROBLEMA 1

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Abre imagen
img = cv2.imread("paris.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Imagen Original")
plt.show()
```

```python
# dimensiones
alto = img.shape[0]
ancho = img.shape[1]

print("Alto :", alto)
print("Ancho:", ancho)
```

```python
#funcion

def activar_color(imagen, color):

    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    imagen_salida = np.zeros((alto, ancho, 3), dtype=np.uint8)
    for i in range(alto):
        for j in range(ancho):
            pixel = imagen[i, j]
            rojo = pixel[0]
            verde = pixel[1]
            azul = pixel[2]
            # Solo azul
            if color == 1:
                imagen_salida[i, j] = [0, 0, azul]
            # Solo verde
            elif color == 2:
                imagen_salida[i, j] = [0, verde, 0]
            # Solo rojo
            elif color == 3:
                imagen_salida[i, j] = [rojo, 0, 0]
            # Rojo + Verde
            elif color == 10:
                imagen_salida[i, j] = [rojo, verde, 0]
            # Verde + Azul
            elif color == 20:
                imagen_salida[i, j] = [0, verde, azul]
            # Azul + Rojo
            elif color == 30:
                imagen_salida[i, j] = [rojo, 0, azul]

    return imagen_salida
```

```python
#Prueba
azul = activar_color(img,1)
verde = activar_color(img,2)
rojo = activar_color(img,3)
amarillo = activar_color(img,10)
cyan = activar_color(img,20)
magenta = activar_color(img,30)
```

```python
#Resultados

plt.figure(figsize=(14,8))
plt.subplot(2,3,1)
plt.imshow(azul)
plt.title("Solo Azul")
plt.subplot(2,3,2)
plt.imshow(verde)
plt.title("Solo Verde")
plt.subplot(2,3,3)
plt.imshow(rojo)
plt.title("Solo Rojo")
plt.subplot(2,3,4)
plt.imshow(amarillo)
plt.title("Rojo + Verde")
plt.subplot(2,3,5)
plt.imshow(cyan)
plt.title("Verde + Azul")
plt.subplot(2,3,6)
plt.imshow(magenta)
plt.title("Azul + Rojo")

plt.show()
```

```python
#Guarda resltados
cv2.imwrite("solo_azul.jpg", cv2.cvtColor(azul, cv2.COLOR_RGB2BGR))
cv2.imwrite("solo_verde.jpg", cv2.cvtColor(verde, cv2.COLOR_RGB2BGR))
cv2.imwrite("solo_rojo.jpg", cv2.cvtColor(rojo, cv2.COLOR_RGB2BGR))
cv2.imwrite("rojo_verde.jpg", cv2.cvtColor(amarillo, cv2.COLOR_RGB2BGR))
cv2.imwrite("verde_azul.jpg", cv2.cvtColor(cyan, cv2.COLOR_RGB2BGR))
cv2.imwrite("azul_rojo.jpg", cv2.cvtColor(magenta, cv2.COLOR_RGB2BGR))
```

```python
##PROBLEMA 3

img = cv2.imread("paris.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Imagen Original")

plt.show()
```

```python
#funcion

def separar_canales(imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    rojo = np.zeros((alto, ancho, 3), dtype=np.uint8)
    verde = np.zeros((alto, ancho, 3), dtype=np.uint8)
    azul = np.zeros((alto, ancho, 3), dtype=np.uint8)
    for i in range(alto):
        for j in range(ancho):
            pixel = imagen[i, j]
            r = pixel[0]
            g = pixel[1]
            b = pixel[2]
            rojo[i, j] = [r, 0, 0]
            verde[i, j] = [0, g, 0]
            azul[i, j] = [0, 0, b]

    return rojo, verde, azul
```

```python
#ejecuta

canal_rojo, canal_verde, canal_azul = separar_canales(img)
```

```python
#resultados

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(canal_rojo)
plt.title("Canal Rojo")
plt.subplot(1,3,2)
plt.imshow(canal_verde)
plt.title("Canal Verde")
plt.subplot(1,3,3)
plt.imshow(canal_azul)
plt.title("Canal Azul")

plt.show()
```

```python
#guardar imagenes

cv2.imwrite(
    "canal_rojo.jpg",
    cv2.cvtColor(canal_rojo, cv2.COLOR_RGB2BGR))

cv2.imwrite(
    "canal_verde.jpg",
    cv2.cvtColor(canal_verde, cv2.COLOR_RGB2BGR))

cv2.imwrite(
    "canal_azul.jpg",
    cv2.cvtColor(canal_azul, cv2.COLOR_RGB2BGR))
```

```python
#funcion

def calcular_histograma(imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    hist_rojo = np.zeros(256, dtype=int)
    hist_verde = np.zeros(256, dtype=int)
    hist_azul = np.zeros(256, dtype=int)
    hist_gris = np.zeros(256, dtype=int)
    suma_gris = 0
    cantidad_pixeles = alto * ancho
    for i in range(alto):
        for j in range(ancho):
            pixel = imagen[i,j]
            rojo = int(pixel[0])
            verde = int(pixel[1])
            azul = int(pixel[2])
            gris = int((rojo + verde + azul)/3)
            hist_rojo[rojo] += 1
            hist_verde[verde] += 1
            hist_azul[azul] += 1
            hist_gris[gris] += 1
            suma_gris += gris
    media = suma_gris / cantidad_pixeles

    return hist_rojo,hist_verde,hist_azul,hist_gris,media
```

```python
#funcion

def gris_ponderado(imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    imagen_gris = np.zeros((alto, ancho, 1), dtype=np.uint8)
    for i in range(alto):
        for j in range(ancho):
            pixel = imagen[i, j]
            rojo = pixel[0]
            verde = pixel[1]
            azul = pixel[2]
            gris = int((0.299 * rojo) +
                       (0.587 * verde) +
                       (0.114 * azul))
            imagen_gris[i, j] = gris
    return imagen_gris
```

```python
#ejecuta

gris = gris_ponderado(img)
```

```python
#resultado

plt.imshow(gris, cmap="gray")
plt.title("Escala de Grises Ponderada")
plt.show()
```

# ¿Qué es el espacio de color HSV?

El espacio de color **HSV** es una forma diferente de representar los colores.

Mientras que el modelo **RGB** define un color indicando cuánto rojo, verde y azul contiene un píxel, el modelo **HSV** describe el color utilizando tres características que se relacionan mejor con la percepción humana.

HSV significa:
- **H (Hue):** Matiz
- **S (Saturation):** Saturación
- **V (Value):** Valor o Brillo


# 1. Hue (Matiz)
El **matiz** indica el color principal.
Se representa mediante un ángulo entre **0° y 360°**.
Algunos ejemplos son:

| Color    | Hue |
|----------|:----:|
| Rojo     | 0°   |
| Amarillo | 60°  |
| Verde    | 120° |
| Cian     | 180° |
| Azul     | 240° |
| Magenta  | 300° |

Cuando el ángulo llega a **360°**, vuelve nuevamente al rojo.

# 2. Saturation (Saturación)
La **saturación** representa la intensidad o pureza del color.
- **0 %:** el color pierde intensidad y se aproxima al gris.
- **100 %:** el color se observa completamente puro.

Por ejemplo:
- Un rojo muy intenso tiene alta saturación.
- Un rojo opaco tiene baja saturación.

# 3. Value (Valor)
El **valor** representa el brillo del color.
- **0 %:** negro.
- **100 %:** brillo máximo.
Un mismo color puede verse más claro o más oscuro modificando únicamente este componente.

# Representación del modelo HSV
- **Hue** → Color
- **Saturation** → Intensidad del color
- **Value** → Brillo


# Diferencias entre RGB y HSV

|                     RGB                                   |                                      HSV                                        |
|-----------------------------------------------------------|---------------------------------------------------------------------------------|
| Representa cantidades de rojo, verde y azul.              | Representa color, saturación y brillo.                                          |
| Es adecuado para capturar y mostrar imágenes.             | Es útil para analizar y segmentar colores.                                      |
| Cambiar un color puede requerir modificar varios canales. | Es posible modificar el tono, la saturación o el brillo de forma independiente. |


# Aplicaciones del espacio HSV
El modelo HSV es ampliamente utilizado en visión por computadora debido a que facilita el trabajo con colores. Algunas aplicaciones son:

- Detección de objetos por color.
- Seguimiento de objetos.
- Segmentación de imágenes.
- Reconocimiento de señales.
- Sistemas de inspección industrial.
- Robótica y navegación basada en visión.

# Ejemplo de mapeo de colores

| Color    | Hue  | Saturation | Value |
|----------|:----:|:----------:|:-----:|
| Rojo     | 0°   | 100 %      | 100 % |
| Verde    | 120° | 100 %      | 100 % |
| Azul     | 240° | 100 %      | 100 % |
| Amarillo | 60°  | 100 %      | 100 % |
| Blanco   | 0°   | 0 %        | 100 % |
| Gris     | 0°   | 0 %        | 50 %  |
| Negro    | 0°   | 0 %        | 0 %   |


# Conclusión

El espacio de color **HSV** representa los colores mediante su tono (**Hue**), saturación (**Saturation**) y brillo (**Value**), lo que facilita la manipulación y el análisis de imágenes en tareas de visión por computadora. A diferencia del modelo **RGB**, HSV permite separar la información del color de la iluminación, siendo especialmente útil para aplicaciones como la detección y segmentación de objetos por color.
